```python
# stack one bar on top of another

batsmen = ['ABD', 'Warner', 'Dhoni', 'Rohit', 'Virat']
runs_2015 = [500, 450, 600, 480, 550]
runs_2016 = [520, 480, 300, 550, 700]
runs_2017 = [600, 500, 200, 650, 800]

x_values = np.arange(len(batsmen))

plt.bar(batsmen, np.array(runs_2017)+np.array(runs_2015)+np.array(runs_2016), label="runs in 2017")
plt.bar(batsmen, np.array(runs_2015)+np.array(runs_2016), label="runs in 2016")
plt.bar(batsmen, runs_2015, label="runs in 2015")
plt.xticks(rotation=45)
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
seasons = np.arange(2012, 2021)
price = [45000, 48000, 50000, 54000, 60000, 65000, 70000, 75000, 120000]

# Create figure and axes
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Matplotlib Comprehensive Demo', fontsize=16)

# 1. Line plot
axs[0,0].plot(seasons, price, marker='s', linestyle='--', color='teal', linewidth=2)
axs[0,0].set_title('iPhone Price Trend')
axs[0,0].set_ylabel('Price (INR)')
axs[0,0].grid(True, linestyle=':', alpha=0.7)

# 2. Scatter
x = np.random.randn(100)
y = 2*x + np.random.randn(100)*0.5
colors = np.random.rand(100)
sizes = np.random.randint(10, 100, 100)
sc = axs[0,1].scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
axs[0,1].set_title('Scatter with Colormap & Varying Size')
fig.colorbar(sc, ax=axs[0,1])

# 3. Bar
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
axs[1,0].bar(categories, values, color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axs[1,0].set_title('Bar Chart')

# 4. Histogram
data = np.random.normal(60, 15, 500)
axs[1,1].hist(data, bins=20, color='purple', alpha=0.7, edgecolor='white')
axs[1,1].axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2)
axs[1,1].set_title('Histogram with Mean Line')

plt.tight_layout()
plt.savefig('combined_plots.png', dpi=150)
plt.show()
```
