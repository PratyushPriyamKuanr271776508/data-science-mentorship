```python
%matplotlib inline
import matplotlib.pyplot as plt

year = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
price = [45000, 48000, 50000, 54000, 60000, 65000, 70000, 75000, 120000]

plt.plot(year, price)
plt.title('Car Price Trend')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(year, price, linestyle='--')   # dashed
plt.plot(year, price, linestyle=':')    # dotted
plt.plot(year, price, linestyle='-.')   # dash-dot
```

```python
plt.plot(year, price, label="iphone Price", marker="o")
plt.xlabel("Year")
plt.ylabel("Price(INR)")
plt.title("iPhone price over years")
plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.plot(year, price, label="iphone Price", marker="o")
plt.xlabel("Year")
plt.ylabel("Price(INR)")
plt.title("iPhone price over years")
plt.legend()
plt.ylim(0, 100000)   # cuts off the 1.2L outlier, shows variation clearly
plt.xlim(2017, 2019)  # zoom into a specific time range
plt.grid(True)
plt.show()
```

```python
seasons = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']
rohit_runs = [0, 362, 401, 372, 433, 538, 430, 482, 489, 333]
virat_runs = [0, 165, 307, 557, 364, 634, 359, 505, 973, 308]

plt.plot(seasons, rohit_runs, label="ROHIT IPL RUNS", color="blue", marker='.')
plt.plot(seasons, virat_runs, label="VIRAT IPL RUNS", color="green", marker='*')

plt.xlabel('SEASONS')
plt.ylabel('TOTAL RUNS')
plt.title("VIRAT VS ROHIT TOTAL SCORES PER SEASON")
plt.grid(True)
plt.legend()
plt.show()
```
